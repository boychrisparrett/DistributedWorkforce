```python
from BaseOrg import *
from EmployeeIDs import *
from WorldClock import *
from ShiftTypes import *
import datetime
import matplotlib.pyplot as plt
%matplotlib inline 

class PanamaOrg(BaseOrg):
    def __init__(self,uid,n_emps,empids,clock,shfttypes):
        super().__init__(uid,n_emps,empids,clock,shfttypes)
        
    def _LoadEmployees_(self):
        #regular... overload function for different
        cur_t = self.wrldclock.getDTG()
        shifts = ["Panama_Team1_AM","Panama_Team1_PM","Panama_Team2_AM","Panama_Team2_PM"]
        for s in shifts:
            print("Loading shift", s)
            for n in range(int(self.numemps/4)):
                nid = self.availempids.getEID()
                print("\t emp:", nid)
                newagt = BaseAgent(nid,self.shifttypes.getShift(s),cur_t)
                newagt.org = self.uid
                self.members[nid] = newagt
                self.stats[nid] = [0]


MStartTime = datetime.datetime(year=2000,month=1,day=1,hour=0, minute=0)
TimeStep = 60 #minutes per step
GlblEIDS = EmployeeIDs(100,10000)
GlblClock = WorldClock(TimeStep,epoch=MStartTime) 
GlblShift = ShiftTypes(GlblClock)
GlblShift.LoadDefault()
NumEmps = 20
mybase = BaseOrg(1,NumEmps,GlblEIDS,GlblClock,GlblShift)
mypanm = PanamaOrg(2,NumEmps,GlblEIDS,GlblClock,GlblShift)
mycomp = BaseOrg(3,NumEmps,GlblEIDS,GlblClock,GlblShift,stype="Compressed")

for n in range(14*24):
    GlblClock.Step()
    mybase.step()
    mypanm.step()
    mycomp.step()
```

```python
ts = np.arange('2000-01-01', '2000-01-15', dtype='datetime64[h]')

memkeys = list(mybase.stats.keys())
basear = mybase.stats[memkeys[0]]
for n in memkeys:
    basear = np.add(basear,mybase.stats[n])

memkeys = list(mycomp.stats.keys())
comppar = mycomp.stats[memkeys[0]]
for n in memkeys:
    comppar = np.add(comppar,mycomp.stats[n])

memkeys = list(mypanm.stats.keys())
nempar = mypanm.stats[memkeys[0]]
for n in memkeys[1:]:
    nempar = np.add(nempar,mypanm.stats[n])
    
plt.plot(ts,comppar[1:])
plt.plot(ts,nempar[1:])
plt.plot(ts,basear[1:])
```
